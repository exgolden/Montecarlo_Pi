```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N=int(input("Ingrese el numero de elementos en base 10: "))
```

```python
n=100000
# Creamos las coordenadas aleatorias
X=(np.random.random(n))
Y=(np.random.random(n))
# Centramos las coordenadas en (0,0)
X=X-0.5
Y=Y-0.5
```

```python
# Creamos el plot para visualizar los elementos aleatorios
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(X, Y, color="yellow", s=1)
circle=plt.Circle((0, 0), 0.5, color="red", fill=False, linewidth=3)
square=plt.Rectangle((-0.5, -0.5), 1, 1, color="blue", fill=False, linewidth=3)
ax.set_facecolor("#808080") 
ax.set_aspect("equal")
ax.add_patch(circle)
ax.add_patch(square)
plt.xticks([])
plt.yticks([])
plt.tick_params(top="off", bottom="off", left="off", right="off",
                labelleft="off", labelbottom="on")
```

```python
# Creamos una matriz de tama√±o Nx2 y elevamos al cuadrado
X_Y=np.column_stack((X, Y))
X_Y=X_Y**2
# Contamos el numero de elementos dentro del radio
Count=0
r2=0.5**2
for i in range(len(X_Y)):
    if (np.sum(X_Y[i])<r2):
        Count+=1
# Aproximamos usando el ratio
pi=4*(Count/n)
pi
```
