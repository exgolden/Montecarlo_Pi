 # Integracion de Monte Carlo

Partimos de la formula del area para un circulo(C) y un cuadrado(S):
$$Area_{C}=\pi r^2$$
$$Area_{S}= l^2$$

Ahora escogamos *r=1*, para tener un circulo unitario y un cuadrado con lados de longitud *l=2r*, de manera que nuestro circulo unitario quede dentro del cuadrado, y tengamos:

$$\frac{Area_{C}}{Area_{S}}=\frac{\pi}{4}$$
y por lo tanto:

$$\pi=4 \left( \frac{Area_{C}}{Area_{S}} \right)$$

Podemos aproximar los digitos de Pi que querramos mediante la integracion de Montecarlo. Sea:

 $$F(x,y)= \left\{
\begin{array}{ll}
      1 & ,x^{2}+y^{2} \leq 1 \\
      0 & ,x^{2}+y^{2} > 1 \\
\end{array} 
\right. $$
y sea $$\Omega=[-1,1] \times [-1,1]$$

entonces
$$4\int_{\Omega} F(x,y) \,dxdy =\pi$$

$$4\int_{\Omega} F(x,y) \,dxdy \approx 4\frac{1}{N}\sum_{n=1}^{N} F(x_{i},y_{i}) \approx \pi$$

Ahora con un algoritmo establecido podemos aproximar Pi

```python
import numpy as np
import decimal
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N=int(input("Ingrese el numero de elementos en base 10: "))
```

```python
# Creamos las coordenadas aleatorias
X=(np.random.random(N))
Y=(np.random.random(N))
# Centramos las coordenadas en (0,0)
X=X-0.5
Y=Y-0.5
    # Creamos una matriz de tama√±o Nx2 y elevamos al cuadrado
X_Y=np.column_stack((X, Y))
X_Y=X_Y**2
# Contamos el numero de elementos dentro del radio
Count=0
r2=0.5**2
for i in range(len(X_Y)):
    if (np.sum(X_Y[i])<r2):
        Count+=1
# Aproximamos usando el ratio
Pi=4*(Count/N)
```

```python
# Creamos el plot para visualizar los elementos aleatorios
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(X, Y, color="#000000", s=1)
circle=plt.Circle((0, 0), 0.5, color="red", fill=False, linewidth=3)
square=plt.Rectangle((-0.5, -0.5), 1, 1, color="blue", fill=False, linewidth=3)
ax.set_facecolor("#FFFFFF") 
ax.add_patch(circle)
ax.add_patch(square)
plt.xticks([])
plt.yticks([])
```
